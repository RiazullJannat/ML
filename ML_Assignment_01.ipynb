# Question 1 (10 Marks)
Load the Titanic dataset and display:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("./sample_data/titanic_data_updated.csv")
print(f"Total samples = {df.shape[0]}")
print(f"Total features = {df.shape[1]}")
print("\nFirst 10 samples:\n")
display(df.head(10))
print("\nRandom 5 samples:\n")
df.sample(5)
```

# Question 2 (10 Marks)
Perform feature screening on the Titanic dataset by:
- Finding total missing values
- Finding duplicate rows
- Removing duplicate rows permanently

```python
print(f"Total missing values = \n")
print(df.isnull().sum())
print(f"\nTotal duplicate rows = {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
```

# Question 4 (10 Marks)
Perform univariate categorical analysis on the Survived column by:
- Creating a countplot
- Calculating percentage distribution
- Creating a pie chart

Write 2 insights from the analysis.

```python
print("Count plot survived column:\n")
sns.countplot(data=df,x="Survived")
plt.show()

print("\n\nPercentage distribution:\n")
survived_counts = df["Survived"].value_counts(normalize=True)*100
print(survived_counts)

print("\n\nPie chart:\n")
plt.pie(survived_counts,labels=survived_counts.index,autopct="%1.1f%%")
plt.show()
```

# Question 6 (10 Marks)
Perform multivariate analysis between Sex and Survived using a countplot with hue.

Also calculate normalized survival ratios using groupby().

Write 2 insights from the analysis.


```python
print("Count plot with hue:\n")
sns.countplot(data=df,x="Sex",hue="Survived")
plt.show()

print("\n\nNormalized survival ratios:\n")

print("\nNormalized Survival Ratio:\n")
print(df.groupby("Sex")["Survived"].value_counts(normalize=True)*100)

print(
    f"""\n2 insights from the analysis:
Understand survived rate between male and female.
Female survived rate is higher than male.
Male survived rate is so poor compared to total male passenger.
"""
)
```
