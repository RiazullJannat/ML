```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = sns.load_dataset('penguins')
print(f"total number of samples: {df.shape[0]}")
print(f"total number of features: {df.shape[1]}")
```

 count plot for the species column

```python
sns.countplot(data=df, x='species')
```

```python
# Calculate the percentage distribution of the 'island' column
island_distribution = df['island'].value_counts(normalize=True) * 100

print("Percentage distribution of 'island' column:")
print(island_distribution)
```

```python
sex_distribution = df['sex'].value_counts(normalize=True) * 100

plt.pie(sex_distribution, labels=sex_distribution.index, autopct='%1.1f%%')
plt.title('Sex Distribution')
plt.show()

```
