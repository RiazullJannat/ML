```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./sample_data/titanic_data_updated.csv")
df.head(5)
```

#Multivariante analysis

```python
sns.countplot(data=df, x="Sex", hue="Survived")
```

```python
gender_split = df.groupby("Sex")['Survived'].value_counts(normalize=True)
print(gender_split)
```

```python
# spliting Pclass
sns.countplot(data=df, x="Pclass", hue="Survived")
pclass_split = df.groupby("Pclass")['Survived'].value_counts()
print(pclass_split)

```

```python
#spliting by Embarked
sns.countplot(data=df, x="Embarked", hue="Survived")
embarked_split = df.groupby("Embarked")['Survived'].value_counts()
print(embarked_split)
#
```

```python
#spliting by age
sns.countplot(data=df, x="Age", hue="Survived")
age_split = df.groupby("Age")['Survived'].value_counts()
print(age_split)
```

```python
sns.barplot(x=df['Survived'],y=df['Age'],hue=df['Sex'])
```

```python
sns.kdeplot(data=df, x="Age", hue="Survived")
```
