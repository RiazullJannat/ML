```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("./sample_data/titanic_data_updated.csv")
df
```

```python
#remove unnecessary columns
df.drop(["PassengerId","Name", "Ticket"], axis=1, inplace=True)
df
```

```python
# spliting features and target columns

features = df.drop("Survived", axis=1)
target = df["Survived"]
```

```python
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size=0.2, random_state=42)
features_test.isnull().sum()
```

```python
#declearing imputer
age_imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
embarked_imputer = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
cabin_imputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value="Missing", add_indicator=True)

#Learning the columns
age_imputer.fit(features_train[['Age']])
embarked_imputer.fit(features_train[["Embarked"]])
cabin_imputer.fit(features_train[['Cabin']])


#transforming the values
features_train['Age'] = age_imputer.transform(features_train[['Age']]).ravel()
features_train['Embarked']=embarked_imputer.transform(features_train[['Embarked']]).ravel()
features_train[["Cabin","cabin_indicator"]]=cabin_imputer.transform(features_train[['Cabin']])


features_test['Age'] = age_imputer.transform(features_test[['Age']]).ravel()
features_test['Embarked']=embarked_imputer.transform(features_test[['Embarked']]).ravel()
features_test[["Cabin","cabin_indicator"]]=cabin_imputer.transform(features_test[['Cabin']])


print("Train data set after clning nulls:\n")
display(features_train)

print("Test data set after clning nulls:\n")
display(features_test)
```
